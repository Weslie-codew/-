```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取 Excel 文件
file_path = '/home/wmy/Desktop/ecommerce_product_recommendation_collaborative.xlsx'
data = pd.read_excel(file_path)

# 1. 年龄分布直方图
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=20, color='blue', edgecolor='black')
plt.title('年龄分布直方图')
plt.xlabel('年龄')
plt.ylabel('用户数量')
plt.grid(True)
plt.show()

# 2. 收入分布箱线图
plt.figure(figsize=(10, 6))
plt.boxplot(data['Income'], vert=False)
plt.title('收入分布箱线图')
plt.xlabel('收入')
plt.grid(True)
plt.show()

# 3. 年龄与购买频率散点图
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Purchase_Frequency'], alpha=0.5, color='green')
plt.title('年龄与购买频率散点图')
plt.xlabel('年龄')
plt.ylabel('购买频率')
plt.grid(True)
plt.show()

# 4. 收入与总消费额散点图
plt.figure(figsize=(10, 6))
plt.scatter(data['Income'], data['Total_Spending'], alpha=0.5, color='red')
plt.title('收入与总消费额散点图')
plt.xlabel('收入')
plt.ylabel('总消费额')
plt.grid(True)
plt.show()

# 5. 产品类别偏好热力图
# 假设 Product_Category_Preference 是一个分类变量，需要先将其转换为数值
data['Product_Category_Preference'] = data['Product_Category_Preference'].astype('category').cat.codes

plt.figure(figsize=(10, 6))
sns.heatmap(data[['Product_Category_Preference']].corr(), annot=True, cmap='coolwarm')
plt.title('产品类别偏好热力图')
plt.show()
```
