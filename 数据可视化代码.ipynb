```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  Excel 
file_path = '/home/wmy/Desktop/ecommerce_product_recommendation_collaborative.xlsx'
data = pd.read_excel(file_path)

# 1. 
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=20, color='blue', edgecolor='black')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()

# 2. 
plt.figure(figsize=(10, 6))
plt.boxplot(data['Income'], vert=False)
plt.title('')
plt.xlabel('')
plt.grid(True)
plt.show()

# 3. 
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Purchase_Frequency'], alpha=0.5, color='green')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()

# 4. 
plt.figure(figsize=(10, 6))
plt.scatter(data['Income'], data['Total_Spending'], alpha=0.5, color='red')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()

# 5. 
#  Product_Category_Preference 
data['Product_Category_Preference'] = data['Product_Category_Preference'].astype('category').cat.codes

plt.figure(figsize=(10, 6))
sns.heatmap(data[['Product_Category_Preference']].corr(), annot=True, cmap='coolwarm')
plt.title('')
plt.show()
```
